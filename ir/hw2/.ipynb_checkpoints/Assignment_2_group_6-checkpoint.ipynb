```python
# Import libraries

# !pip install --user -U nltk
import nltk
#from nltk.tokenize import word_tokenize
from nltk.tokenize import RegexpTokenizer
# nltk.download('punkt')

from nltk.corpus import stopwords 
# nltk.download('stopwords')

from nltk.stem import PorterStemmer

from collections import Counter
from itertools import chain
```

```python
# Read the document
documents_path = './AssociatedPress.txt'
with open(documents_path, 'r', encoding='utf-8') as doc_f:
    corpus_list = doc_f.readlines()
```

```python
# Tokenizing all the articles

tokenizer = RegexpTokenizer(r'\w+')
tokenized_corpus = []
for i in range(len(corpus_list) - 1):
    tokenized_corpus.insert(i, tokenizer.tokenize(corpus_list[i])) # tokenized_corpus[i] -> i-th article
```

```python
# Computing number of distinct terms
total_sum = 0
for i in range(len(tokenized_corpus) - 1):
    total_sum += len(tokenized_corpus[i])
    print("Number of tokens in " + str(i) + "-th article: " + str(len(tokenized_corpus[i])))
```

```python
print("Total number of tokens: " + str(total_sum))
```

```python
# Removing numbers 
digitless_tokens = []
for i in range(len(tokenized_corpus) - 1):
    digitless_tokens.insert(i, [s for s in tokenized_corpus[i] if not s.isdigit()])
```

```python
# Computing number of distinct terms
total_sum = 0
for i in range(len(digitless_tokens) - 1):
    total_sum += len(digitless_tokens[i])
    print("Number of tokens in " + str(i) + "-th article: " + str(len(digitless_tokens[i])))
```

```python
print("Total number of tokens: " + str(total_sum))
```

```python
# Case Folding
casefolded_tokens = []
for i in range(len(digitless_tokens) - 1):
    casefolded_tokens.insert(i, [s.casefold() for s in digitless_tokens[i]])
```

```python
# Computing number of distinct terms
total_sum = 0
for i in range(len(casefolded_tokens) - 1):
    total_sum += len(casefolded_tokens[i])
    print("Number of tokens in " + str(i) + "-th article: " + str(len(casefolded_tokens[i])))
```

```python
print("Total number of tokens: " + str(total_sum))
```

```python
# Removing English stopwords
stop_words = set(stopwords.words('english')) 
stopwordLess_tokens = []
for i in range(len(casefolded_tokens) - 1):
    stopwordLess_tokens.insert(i, [s for s in casefolded_tokens[i] if s not in stop_words])
```

```python
# Computing number of distinct terms
total_sum = 0
for i in range(len(stopwordLess_tokens) - 1):
    total_sum += len(stopwordLess_tokens[i])
    print("Number of tokens in " + str(i) + "-th article: " + str(len(stopwordLess_tokens[i])))
```

```python
print("Total number of tokens: " + str(total_sum))
```

```python
# Stemming
ps = PorterStemmer()
stemmed_tokens = []
for i in range(len(stopwordLess_tokens) - 1):
    stemmed_tokens.insert(i, [ps.stem(s) for s in stopwordLess_tokens[i]])
```

```python
# Computing number of distinct terms
total_sum = 0
for i in range(len(stemmed_tokens) - 1):
    total_sum += len(stemmed_tokens[i])
    print("Number of tokens in " + str(i) + "-th article: " + str(len(stemmed_tokens[i])))
```

```python
print("Total number of tokens: " + str(total_sum))
```

```python
# Computing the frequencies of distinct terms after Case Folding
all_tokens = []
for i in range(len(casefolded_tokens) - 1):
    all_tokens.append(casefolded_tokens[i])
    
res = dict(Counter(chain.from_iterable(map(set, all_tokens))))
```

```python
# Final list of tokens sorted by frequency
dict(sorted(res.items(), key=lambda item: item[1],reverse=True))
```
